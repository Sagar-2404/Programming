```python
#Time Complexity of different sorting algorithms.
#                      Best      Average      Worst
#Selection Sort         Ω(n^2)    θ(n^2)      O(n^2)
#Bubble Sort            Ω(n)      θ(n^2)      O(n^2)
#Insertion Sort         Ω(n)      θ(n^2)      O(n^2)
#Heap Sort          Ω(n log(n))  θ(n log(n)) O(n log(n))
#Quick Sort         Ω(n log(n))  θ(n log(n))  O(n^2) 
#Merge Sort         Ω(n log(n))  θ(n log(n)) O(n log(n)) 


# Bubble Sort
def bubble(arr):
    n=len(arr)
    a=True
    for i in range(0,n):
        if a==False:
            break
        for j in range(0,n-1):
            if(arr[j]>arr[j+1]):
                arr[j],arr[j+1]=arr[j+1],arr[j]
                a=True
    return arr

#Merge Sort
def merge(arr):
    if len(arr)>1:
        mid=len(arr)//2
        L=arr[:mid]
        R=arr[mid:]
        merge(L)
        merge(R)
        i=j=k=0
        while i<len(L) and j<len(R):
            if L[i]<R[j]:
                arr[k]=L[i]
                i+=1
            else:
                arr[k]=R[j]
                j+=1
            k+=1
            
        while i<len(L):
            arr[k]=L[i]
            i+=1
            k+=1
        while j<len(R):
            j+=1
            k+=1

#Insertion Sort            
def Insertion(arr):
    min1=arr[2]
    minIndex=-1
    n=len(arr)
    for i in range(0,n-1):
        min1=arr[i]
        for j in range(i+1,n):
            if min1>arr[j]:
                min1=arr[j]
                minIndex=j
        if minIndex!=-1:        
            arr[i],arr[minIndex]=arr[minIndex],arr[i]  
    return arr

#Quick sort
def quick(start,end,arr):
    if start < end:
        p=partition(start,end,array)
        quick(start,p-1,arr)
        quick(p+1,end,arr)
def partition(start, end, array):
    pi_index=start
    pivot=array[pi_index]
    while start<end:
        while start<len(array) and array[start]<=pivot:
            start+=1
        while array[end]>pivot:
            end-=1
        if start<end:
            array[start],array[end]=array[end],array[start]
    array[end],array[pi_index]=array[pi_index],array[end]
    return end

#Count sort
def count_sort(arr):
    max_element= int(max(arr))
    min_element= int(min(arr))
    range_=max_element-min_element+1
    
    count_arr=[0 for _ in range(range_)]
    output_arr=[0 for _ in range(len(arr))]
    
    for i in range(0,len(arr)):
        count_arr[arr[i]-min_element]+=1
    
    for i in range(1,len(count_arr)):
        count_arr[i]+=count_arr[i-1]
        
    for i in range(len(arr)-1,-1,-1):
        count_arr[count_arr[arr[i]-min_element]-1]=arr[i]
        count_arr[arr[i]-min_element]-=1
        
    for i in range(0,len(arr)):
        arr[i]=output_arr[i]
    return arr

# Driver program to test above functions
arr2 = [-5, -10, 0, -3, 8, 5, -1, 10]
ans = count_sort(arr2)
print("Count Sort" ,ans)
array = [10, 7, 8, 9, 1, 5 ]
quick(0, len(array) - 1, array)
print("Quick Sort",array)
arr=[5,4,3,2,1]
a=bubble(arr)
print("Bubble Sort",a)
print("Insertion Sort",Insertion(arr))
merge(arr)
print("Merge Sort",arr)
```
