```python
#Implimentation of max heap.
#Note:If any question such as find the kth maximum element or kth minimum element then use heap data structure.

import sys
class Heap:
    def __init__(self,maxsize):
        self.maxsize=maxsize
        self.size=0
        self.heap=[0]*(maxsize+1)
        self.heap[0]=sys.maxsize
        self.front=1
    def parent(self,pos):
        return pos//2
    def left(self,pos):
        return 2*pos
    def right(self,pos):
        return 2*pos+1
    def isleaf(self,pos):
        if pos>=(self.size//2) and pos<self.size:
            return True 
        return False
    def swap(self,fpos,epos):
        self.heap[fpos],self.heap[epos]=self.heap[epos],self.heap[fpos]
    def insert(self,data):
        if self.size>=self.maxsize:
            return 
        self.size+=1
        self.heap[self.size]=data
        curr=self.size
        while self.heap[curr]>self.heap[self.parent(curr)]:
            self.swap(curr,self.parent(curr))
            curr=self.parent(curr)
    def Print(self):
        for i in range(1,(self.size//2)+1):
            print("Parent:",self.heap[i])
            print("Left child:",self.heap[2*i])
            print("Right child:",self.heap[2*i+1])
            
    def heapify(self,pos):
        if not self.isleaf(pos):
            if self.heap[pos]<self.heap[self.left(pos)] or self.heap[pos]<self.heap[self.right(pos)]:
                if self.heap[self.left(pos)]>self.heap[self.right(pos)]:
                    self.swap(pos,self.left(pos))
                    self.heapify(self.left(pos))
                else:
                    self.swap(pos,self.right(pos))
                    self.heapify(self.right(pos))
        
        
    def extractmax(self):
        popped=self.heap[self.front]
        self.heap[self.front]=self.heap[self.size]
        self.size-=1
        print(self.size)
        self.heapify(self.front)
        return popped            
        

            
heap=Heap(10)
heap.insert(5)
heap.insert(8)
heap.insert(6)
heap.insert(1)
heap.insert(16)
heap.Print()
print("Max",heap.extractmax())
heap.Print()
heap.insert(32)
print("3")
heap.Print()
        
            
```
